A partir de estos datos, el laboratorio desea:

Ajustar una l√≠nea de calibraci√≥n (recta) a los datos.
Utilizar esta l√≠nea para estimar la concentraci√≥n de una soluci√≥n desconocida que tiene una absorbancia de 0.40.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Problema 1: Funci√≥n que calcula la molaridad de una soluci√≥n
def calcular_molaridad(gramos, volumen_ml, peso_molecular):
    moles = gramos / peso_molecular
    volumen_litros = volumen_ml / 1000
    molaridad = moles / volumen_litros
    return molaridad

# Problema 2: Funci√≥n que calcula el pH de una soluci√≥n de √°cido fuerte
def calcular_ph(concentracion_H):
    ph = -np.log10(concentracion_H)
    return ph

# Problema 3: Funci√≥n que calcula la masa molar de un gas ideal
def calcular_masa_molar_gas(gramos, volumen_ml, temperatura_C, presion_atm):
    R = 0.0821  # constante de los gases ideales en L¬∑atm/(mol¬∑K)
    temperatura_K = temperatura_C + 273.15
    volumen_L = volumen_ml / 1000
    masa_molar = (gramos * R * temperatura_K) / (presion_atm * volumen_L)
    return masa_molar

# Problema 4: Funci√≥n que calcula la concentraci√≥n de √°cido y base conjugada para un buffer
def preparar_buffer(ph_deseado, Ka, volumen_ml_deseado):
    H_concentracion = 10**(-ph_deseado)
    base_conjugada = H_concentracion / Ka
    acido = 1 / (1 + base_conjugada)
    base_conjugada = base_conjugada / (1 + base_conjugada)
    return acido * volumen_ml_deseado, base_conjugada * volumen_ml_deseado

# Problema 5: Funci√≥n que calcula la energ√≠a liberada en una reacci√≥n qu√≠mica
def calcular_energia_liberada(moles_reactivo, delta_H_formacion):
    energia = moles_reactivo * delta_H_formacion
    return energia

# Problema 6: Funci√≥n que calcula el rendimiento porcentual
def calcular_rendimiento_porcentual(masa_teorica, masa_experimental):
    rendimiento = (masa_experimental / masa_teorica) * 100
    return rendimiento

# Problema 7: Funci√≥n que calcula el calor espec√≠fico de un metal
def calcular_calor_especifico(masa, delta_T, calor):
    calor_especifico = calor / (masa * delta_T)
    return calor_especifico

# Problema 8: Funci√≥n que calcula la presi√≥n osm√≥tica a 25¬∞C
def calcular_presion_osmotica(concentracion_g_L, peso_molecular, temperatura_C):
    R = 0.0821  # constante de los gases ideales en L¬∑atm/(mol¬∑K)
    moles_soluto = concentracion_g_L / peso_molecular
    temperatura_K = temperatura_C + 273.15
    presion_osmotica = moles_soluto * R * temperatura_K
    return presion_osmotica

# Problema 9: Funci√≥n que calcula el producto de solubilidad de una sal insoluble
def calcular_producto_solubilidad(solubilidad):
    producto_solubilidad = solubilidad ** 2
    return producto_solubilidad

# Problema 10:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress




# Aplicaci√≥n de las funciones con los datos espec√≠ficos

# Problema 1
molaridad = calcular_molaridad(10, 500, 58.44)
print(f"Molaridad de la soluci√≥n: {molaridad} M")

# Problema 2
ph = calcular_ph(0.01)
print(f"pH de la soluci√≥n √°cida: {ph}")

# Problema 3
masa_molar_gas = calcular_masa_molar_gas(2, 1000, 25, 1)
print(f"Masa molar del gas ideal: {masa_molar_gas} g/mol")

# Problema 4
acido, base_conjugada = preparar_buffer(4.5, 1.8e-5, 100)
print(f"Concentraci√≥n de √°cido: {acido} ml, Concentraci√≥n de base conjugada: {base_conjugada} ml")

# Problema 5
energia = calcular_energia_liberada(2, -286)
print(f"Energ√≠a liberada en la reacci√≥n: {energia} kJ")

# Problema 6
rendimiento = calcular_rendimiento_porcentual(5, 4.5)
print(f"Rendimiento porcentual de la reacci√≥n: {rendimiento}%")

# Problema 7
calor_especifico = calcular_calor_especifico(50, 30, 600)
print(f"Calor espec√≠fico del metal: {calor_especifico} cal/g¬∑¬∞C")

# Problema 8
presion_osmotica = calcular_presion_osmotica(10, 60.05, 25)
print(f"Presi√≥n osm√≥tica de la soluci√≥n: {presion_osmotica} atm")

# Problema 9
producto_solubilidad = calcular_producto_solubilidad(0.001)
print(f"Producto de solubilidad de la sal: {producto_solubilidad}")

# Problema 10

# Datos de concentraci√≥n y absorbancia
concentraciones = np.array([0, 10, 20, 30, 40, 50])
absorbancias = np.array([0.0, 0.15, 0.30, 0.45, 0.60, 0.75])

# Graficar los datos de calibraci√≥n
plt.figure(figsize=(10, 6))
plt.scatter(concentraciones, absorbancias, color='blue', label='Datos de calibraci√≥n')
plt.xlabel('Concentraci√≥n (mg/L)')
plt.ylabel('Absorbancia')
plt.title('Curva de Calibraci√≥n de Absorbancia vs. Concentraci√≥n')

# Ajustar una l√≠nea recta a los datos
slope, intercept, r_value, p_value, std_err = linregress(concentraciones, absorbancias)
plt.plot(concentraciones, slope * concentraciones + intercept, color='red', label=f'Ajuste lineal (y = {slope:.2f}x + {intercept:.2f})')
plt.legend()
plt.grid(True)
plt.show()

# Determinar la concentraci√≥n de una soluci√≥n desconocida con absorbancia de 0.40
absorbancia_desconocida = 0.40
concentracion_desconocida = (absorbancia_desconocida - intercept) / slope
print(f"La concentraci√≥n de la soluci√≥n desconocida es aproximadamente: {concentracion_desconocida:.2f} mg/L")
```
